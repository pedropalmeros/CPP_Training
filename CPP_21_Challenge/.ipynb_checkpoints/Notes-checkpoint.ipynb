# Uso del constructor de copia

Además de proporcionar un constructor y un destructor predeterminados, el compilador proporciona un constructor de copia predeterminado. este constructor se llama cada vez que se crea la copia de un objeto. 

Todos los constructores de copia toman como parámtetro una referencia  a un objeto de la misma clase. Es una buena idea hacerla una referencia constante porque el constructor no tendrá que alterar el objeto que se está pasando. 

Ejemplo: 

````c++
GATO(const GATO & elGato);

CLASS(const CLASS & object); 
````

En este caso el constructor `GATO` toma una referencia constatne a un objeto `GATO` existente. El objetivo del constructor de copia es crear una copia de `elGato`.

El constructor de copia predeterminado simplemente copia cada variable miembro del objeto que se pasa como parámetros a las variables miembro del nuevo objeto. Esto se conoce como *copia de los datos miembro (o superficial),* y aunque esto está bien para la mayoría de las variables miembro, no es adecuado para las variables miembro que son apuntadores a objetos que se encuentran en el heap. 

Una copia superficial ode los datos miembro copia en otro objeto los valores exactos de las variables miembro de un objeto. Los apuntadores de ambos eobjetos terminan apuntando a la misma dirección de memoria. Una copia profunda copia en memoria recién asignado los valoress asignados en el heap. 

La solución es crear un propio constructor de copia y asingalr la memoria según sus necesidades.  a Esto se le conoce com copia profunda. 

# Sobrecarga de operadores

## Sobrecarga de operadore prefijo
Los operadores de prefijo se pueden sobrcargar al eclarar funciones con la siguiente forma: 
````c++
tipoDeRetorno operator op (parametros)
````
Aquí `op`e s el operador a sobrecargar. Por lo tanto el operador `++`se puede sobrecargar con la siguiente sintaxis: 
````c++
void operator++()
````

## Sobrecarga de operadores posfijo
Por convención, para la declaración del oeprador se proporciona una variable de tipo entero como parátmetro. El valor del parámetro se ginora; sólo es undicación de que es el perador posfijo. 

`const contador & operator++() ;` - prefijo

`const contador operator++(int);` - posfijo

There are some important differences to be taken into account. 
````C++
// prefix 
const Counter & Counter::operator++(){
    ++lVal;
    return *this;
}

// posfix
const Counter Counter::operator++(int){
    Counter temp(*this);
    ++lVal;
    return temp;
}
````

In the posfix the next steps are performed.
1. The dummy `int` parameter distinguishes this from the pre-increment.
2. It copies the current object into `temp`.
3. It increments the current object (`++lVal`)
4. It returns the old value(the copy stored in `temp`).

## Uso del operador suma

El operador incremente es un operador unario. Funciona sólo con un objeto. El operador suma (`+`) es un operador binario, en el que se involucran dos objetos.

El objetivo es poder declarar dos variables  y luego sumarlas. 

## Uso del operador de asignación 

La última función que proporciona el compilador, si usted no especifica una, es el perador de asignación `operator=()`. Este perador se llama siempre que se asigna algo a un objeto. Por ejemplo: 

````C++
GATO gatoUno(5,7); 
GATO gatoDos(3,4); 
// ... aquí puede ir más código
gatoDos = gatoUno; 
````

Aquí se crea `gatoUno`y se inicializa con `suEdad`igual a 5 y `suPeso`igual a 7. Después se crea `gatoDos`y se le asignan los valores 3 y 4. Después se los valores contenidos en `gatoUno` se asignan a `gatoDos`. Aquíi surgen dos custiones ¿Qué pasa si `suEdad`es un apuntador y qué pasa con los valores originales contenidos en `gatoDos`? 

Una copia superficial sólo copia los miembros, y ambos objetos terminanan apuntadno a la misma área de memoria. Una copia profunda, asigna la memoeria necesaria.
