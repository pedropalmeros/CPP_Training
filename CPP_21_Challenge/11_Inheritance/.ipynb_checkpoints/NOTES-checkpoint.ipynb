# Herencia

cuando decimos que algo es un ttipo de otro algo, damos a entender que es una especialización de un orden mayor.

Cualquier obtejo que sea del tip es un tiene algunas características de la descripción de orden mayor. 


## Herencia y derivación 

El lenguaje C++ intenta representar las relaciones permitiendo que haya clases que se deriven de otras clases. La derivación es una forma de expresar la relación `es un`. Usted dreiva una nueva lcase llamada `Perro` de la clase `Mamifero`. 

Se dice que una clase que agrega nueva funcionalidad a una clase existente se deriva de esa clase original. Se dice que la clase original es la clase base de la nueva clase. 

## Sintaxis de la derivación 

Al declarar una clase, puede indicar de qué clase se deriva escribiendo el símbolo de dos puntos (`:`) después el nombre de la clase, seguido del tipo de derivación (público u otro) y de la clase que se deriva. 

````C++
class Perro: public Mamiero
````

````C++
enum RAZA { GOLDEN, CAIRN, DANDIE, SHEETLAND, DOBERMAN, LAB};

class Mammal{
    public:
        //constructor
        Mammal(){}
        ~Mammal(){}

        //access methods
        int getAge() const; 
        void setAge(int);
        int getWeight() const; 
        void setWeight(int); 

        // Other methods
        void Talk() const;
        void Sleep() const; 

    protected:
        int lAge; 
        int lWeight; 
};

class Dog : public Mammal{
    public: 
        // Constructors
        Dog(){}
        ~Dog(){}

        // Access Methods. 
        RAZA getRaza() const; 
        void setRaza (RAZA aRaza); 

        // Other methods.
        void moveTail(); 
        void askFood(); 
    private:
        RAZA lRaza;
};

````

La clase `Dog` hereda de `Mammal`, Cada objeto ` Dog` tendr'a tres vriables miembro `lRaza`, `lWeight` y `lAge`. Observe que la declaraci'on de clase para `Perro` no include las variables miembro `lAge` y `lWeight`. Los objetos `Dog`  heredan estas variables de la clase `Mammal`, junto ccon todos los m'etodos de `Mammal`, excepto el operador de copia, los construcotres y el destructor. 

## Comparación entre privado y protegido. 

Los miembros privados no están disponibles para las clases derivadas.

Los datos y funciones miembros protegidos son completamente visibles para las clases derivdaas, pero en cuanto a lo demás son privados. 

Existen tres especificadores de acceso: público, protegido y privado.
