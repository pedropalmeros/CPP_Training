## Datos miembro estáticos

Hasta ahora probablemente ha pensado que los datos de cada obejto son sólo para ese objeto yo nose comparten entre objetos de una clase. Por ejemplo, si tiene cinco objetos `Gato`, cada uno tiene su propia edad, peso y otros datos. la edad de uno no afecta la edad del otro. 

Sin embargo, a veces necesitará mantener el registro de una reserva de datos. Por ejemplo, tal vez quiera saber cuántos objetos de una clase exiten todavía. las variable miembro estáticas se comparten entre todeas las instancias de una clase. Son un pacto entre los datos globales, que están disponibles para todos los componentes del program, y los datos miembro, que por lo general están dispnible sólo para cada objeto. 

Puede pensar como un miembro estático pertenece a la clase en un lugar de al objeto. Los datos miembro regulares son por objeto, pero los miembros estáticos son uno por clase.
