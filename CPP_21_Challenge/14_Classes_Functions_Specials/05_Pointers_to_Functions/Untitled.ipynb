# Pointers to functions

Let's make an analogy, with arrays to be able to understand the next concept. 

The array's name is a constat pointer to the first element of the aaray, the name of the function is a constant pointer to the function. 

It is possible to declare a pointer variable that points to the function and invoke the function using the pointer. This could be very usefull, it allows you to create programs that decide with functions would be invoked accordingly to the user's action. 

A pointer to a funciton is invoked in the same way as the functions it points to, except that the name of the pointer to the function is used instead of the name of the function.

To assign a punter to a specific function, the functio name has to be assigned without the parenthesis. The function name is a constat pointer to the function itself. Use the pointer to a function of the same way that it would be used the name of the function. The pointer toa funciton must match with the return value and the function signature. 

Example: 
```C++
long (*apFuncUno)(int, int); 
long aFunction(int, int); 
apFuncUno = aFunction; 
apFuncUno(5, 7);
```
