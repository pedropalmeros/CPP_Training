# CMake and Make

The tool `CMake` is an open-source, platform-independent build system. CMake uses text documents denoted as `CMakeLists.txt` files, to manage build environments, like `make`.

## CMakeLists.txt

`CMakeLists.txt` files are simple text configuration files that tell CMake how to build your project. There can be multiple `CMakeLists.txt` files in a project. In fact, one `CMakeLists.txt` file can be included in each directory of the project, indicating how to files in that directory should be build. 

These files can be used to specify the locations of necessary packages, set build flags and environment variables, specify build target names and locations, and other actions. 

## Building your own CMakelists.txt

### Step 1
The first likes that you'll want in your `CMakeLists.txt` are lines that specifies the minimum versions of `cmake` and C++ required to build the project. Add the following lines to your `CMakeLists.txt` and save the file

````
cmake_minimum_required(VERSION 3.11.3)
set(CMAKE_CXX_STANDARD 17)
````
These lines set the minimum cmake version required to 3.11.3 and set the environment variable `CMAKE_CXX_STANDARD` so CMake uses C++ 17

### Step 2
CMake requires that we name the project, so you should choose a name for the porject and then add the following line to `CMakeLists.txt`

````
project(<your_project_name>)
````
### Step 3
Next, we want to add an exxecutable to this project. You can do that with the `add_executable` command by specifying the executable name, alog with the locations of all the source files that you will need. CMake has the ability to automatically find soruce files in a directory, but for now, you can jus specify eac file needed:

````
add_executable(your_executable_name path_to_file_1 path_to_file_2 ... )
````

Hint: The source files you need are the `.cpp` files usually located in `src/` directory. you can specify the path relative to the `CMakeLists.txt` file, so `src/main.cpp` would work, for example.

### Step 4 
A typicial  CMake project will have a build directory in the same place as the top-level `CMakeLists.txt`

You have to create a `build` directory in the `root` folder of the project, and navigate to the new `build` folder, then you can run the next command

````
cmake ..
````
This command directs the `cmake` command at the top-level `CMakeLists.txt` file with `..` This command uses the `CMakeLists.txt` to configure the project and create a `Makefile` in the build directory

Then run `make`

In this command, `make` finds the `Makefile` and uses the instructions in the `Makefile` to build the project.
