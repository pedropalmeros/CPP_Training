# Algorithmic complexity

Algorithmic complexity is a way to reason how much time or memory your code sill use as inputs grow


Bug O notation (O) - It's really just a way to describe how fast or slow your algorithm grows as the problem gets bigger.

Big O isn't about how fast your code runs today on your computer today, it's about how well your code scales to handle bigger and bigger inputs.

## O(1) - Constant Time
It is a constant time, and it is the best case, no matter how your input is, the algorithm always takes the same amount of time whether you vector has 10 elements or 10 million.

## O(n) - Linear Time
Linear time means your algorithm takes longer as the input grows, exactly in proportion, so if it takes 1 second for a thousand items, it will take 2 seconds for 2000 items,.

## O(logn) - Logarithmic Time
An example of this is a binary search, in which you have the item to be found, and the set of opitions is divided in two, and then in two, and so on, until the found item is isolated


## Understanding Algorithm Complexity and Optimizations

Once your algorithm worked, your zoomed out to ask: how efficient is it? This is where Big-O notation comes in. Big-O describes how algorithms scale as input size grows. O(1) means constat time, like accessing an element in a vector. O(n) means linear time, like counter elements in a list. O(log n) describes logarithmic time, such as binary search where each step halves the problem. And O(n log n) often shows up in algorithms like sorting or repeated priority queue operations. More expensive complexities, like O(n^2), come from nested loops and quickly become impractical. Big-O isn't about exact speed --It's about scalability. By understanding it, you learned how to think about performance like a computer scientist.
