# C++ Lambdas

## Introduction

Lambdas are **inline, anonymous functions** you can write right where you use them- perfect for tiny preicates, comparators, and callbacks.

### The shape of a Lambda

````C++
[capture] (parameters) -> return_type{
// body
}
````

- **`[capture]`**: what outside variables the lambda can use
- **`(parameters)`**: just like regular function's parameters.
- **`-> return_type`**: usually optional, (the compiler can deduce it).


### Using Lambdas for Custom Sorting(Comparator)

````C++
#include <algorithm>
#include <vector> 

struct Book {int pages;}; 

int main(){

    std::vector<Book> shelf ={{352},{464}, {128}}; 

    // Sort by pages ascending
    std::sort(shelf.begin(), shelf.end(), [](const Book &a, const Book &b){
        return a.pages < b.pages;
    });

    return 0; 
}
```` 
Here, we're asking the standard algorithm `std::sort` to reorder a vector in **in place**. 

`std::sort` needs two things:

1. a **range** to sort (that's `shelf.begin()` to `shelf.end()`, and
2. a **comparator** - a tiny function that anwswers **should the left item come before the right item**?

In this code, the comparator is a **lambda**:

````C++
[](const Book &a, const Book &b){
    return a.pages < b.pages;
}
````
* `[]`means "capture nothing"- the lambda doesn't rely on any outside variables
* `(const Book& a, const Book& b)` are tow items being compared during sorting.
* The body returns `true` if `a` should come before `b`. Here, that's "fewer pages first¨, so the result is **ascending by** `pages`.




### Using Lambdas to apply an action to elements

````C++
#include <algorithm>
#include <vector>
#include <iostream>

using namespace std; 

int main(){
    vector<int> v = {1,2,3}; 

    std::for_each(v.begin(), v.end(),[](int x){std::cout << x << " "; });
}
````

`std::for_each` walks through a range and **runs a function on each element**. Here, the function is a lambda that prints each number. 
* Again,  `[]` means the lambda doesn´t use anything from outside.
* The parameters `int x`**by value**, so it's a copy for each element. That's fine because we're just printing.
* `std::for_each(v.begin(), v.end(), ...)` simply applies that lambda to every element from start to finish.

What to remember:
* If you want to **modify** the elments in the vector, take them **by reference**: `[](int& x){x *= 2}`.
* If you need to **count** or build a string while you go, **capture** a variable from outside, e.g. `[&count]` or `[&out]`, and update it inside the lambda.
* A range-based for loop is equivalent for simple cases. `std::for_each` shines when you want to **compose** with other algorithms or pass the callable around.


In both snippets, the lambda is just a tiny, local function you define at the call site. With `std::sort`, it defines the **ordering** with `std::for_each`, it defines the **ordering**, with `std::for_each`, it defines the **action**. That's the core idea

### Lambdas Essentials
````
[]            // capture nothing
[x, &y]       // catpure x by value, y by reference
[=]           // capture everything you use by value
[&]           // capture everything you use by reference
[=, &y]       // default by value, but y by reference
[&, x]        // defaultt by reference, but x by value
````
