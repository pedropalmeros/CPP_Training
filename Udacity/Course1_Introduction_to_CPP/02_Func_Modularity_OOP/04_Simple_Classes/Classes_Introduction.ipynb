#  Introduction 

In c++, a class lets you bundle together data and the functions that act on that data. It is the heart of object-oriented programming, and it helps you keep your code organized, maintainable, and safe. Encapsulation means we hide the internal details of an object and expose only the safe ways to interact with it.


In C++, we use private to hide vriables and public to expose functions. 

Think of classes as a mold, and each object you create from it is a copy of that mold with its own values. Each class contains two primary components, memeber variables, also called attributes, where the data is stored, and member functions, also called methods, which define the behaviors it can perform. Oftentimes, classes also contain constructors, which are special functions that initialize the object.
