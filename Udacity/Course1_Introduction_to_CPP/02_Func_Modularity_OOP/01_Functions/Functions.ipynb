# Passing by Value

````c++
void reset(int x){
    x = 0; 
}

int main(){
    int score = 99; 
    reset(score); 
    std::cout << score << std::endl; 
}
````

# Passing by Reference

````c++
void reset(int &x){ // &tells the compiler to pass the parameter as a reference
    x = 0; 
}

int main(){
    int score = 99; 
    reset(score); 
    std::cout << score << std::endl; 
}
````

# Passing by Pointer

````c++
void reset(int* x){
    if(x != nullptr){
        *x = 0; 
    }
}

int main(){
    int score = 99; 
    reset(&score); 
    std::cout << score << std::endl; 
}
````
