# Passing by Value

````c++
void reset(int x){
    x = 0; 
}

int main(){
    int score = 99; 
    reset(score); 
    std::cout << score << std::endl; 
}
````

# Passing by Reference

````c++
void reset(int &x){
    x = 0; 
}

int main(){
    int score = 99; 
    reset(score); 
    std::cout << score << std::endl; 
}
````
