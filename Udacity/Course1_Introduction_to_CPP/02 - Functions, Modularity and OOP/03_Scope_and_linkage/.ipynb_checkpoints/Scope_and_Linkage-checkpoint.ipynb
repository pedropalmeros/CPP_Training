# Scope and Linkage

## Linkage

Linkage controls whether functions and variables are visible only within one flie or shared across multiple files. By default, most global variables and functions in C++ have external linkage

**EXTERNAL LINKAGE** when a variable or function has external linkage means that they can be accessed by other files during the linking phase. 

---
utils.cpp
````c++
int count = 0; 

// code 
````
---

main.cpp
````c++
extern int count; 

// code
````
---
 In the previous code, the variable `count` is defined in the `utils.cpp` but it can be accessed in main, because it has the key word `extern` creating the external linkage. The `extern` syntax tells the compiler that cout exists somewhere else, don't define it again. 

 Functions also have external linkage bydefault, but you don't need extern for those unless you want to be explicit. Sometimes we don't want a variable or function shraed accross files. To keep it private to the file as defined in, you can give it internal linkage using the `static` keyword

`logger.cpp`
