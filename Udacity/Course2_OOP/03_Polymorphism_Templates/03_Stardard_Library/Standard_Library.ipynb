**Predicates** are functions that return true or false. The standard library algorithms use a predicate to decide which elements match a condition. Or example, use count if to count matches, find if to find the first match, and  any of, all of, or non of to answer existence questtions. Use remove If to filter a container. use captures to parameterize the test. When you capture by value, you make a copy of the variable at the moment the Lambda is created. For example, if min age is 18, when you build the predicate, the predicate will keep using 18, even if you later change min age to 21 before the algorithm runs. When you capture by referece, the Lambda reads the original variable. You change the variable later, the predicate sees the new value. Stays alive means the reference variable mus still exist for as long as the Lambda might run. This is safe when you run the algorithm immediately in the same scope. it is unsafe if you store the lambda and use it after the variable ahs one out of scope because the reference would point to a destroyed object. When in doubt, prefer capturing by value for small copyable settings like thresholds and flags. when filtering with remove if, remember that it does not erase elements by itself, It moves the unwanted elements to the end, and then you erase that tail iwht the containers erase method.
