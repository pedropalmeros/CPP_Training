# Exercise - Polymporphic Notifier

## Scenario
You're wiring up two systems messages for a new whatsapp: 

- a **welcome** message for new users
- a **2FA code** confirmation vis SMS

Different channels (email vs SMS) should be swappeable behind a single interface so other parts of the app don't care *how* the message is sent - just that it **is** sent. This is a perfect use of an abstract interface + runtime dispatch via **base-class reference** (no pointers). 

## Task

- Create an abstracct base class **Notifier** with a pure virtual  `void send(const std::string& msg) const = 0;`
- Implement two concrete classes:
    - EmailNofifier -> prints Email:
    - SmsNotifier -> prints SMS:
- Write a free function void broadcast(const Notifier&n, const std::string& msg) that calls n.send(msg)(**base-class reference**; still polymorphic, no pointers).
- In main() (no input), create a EmailNotifier and SmsNotifier on the stack and call:
    - broadcast(email,"Welcome aboard!)
    - broadcast(sms, "Code 1234 confirmed")
Success criteria
- Uses an **abstract class** + **override**.
- Uses **runtime dispatch** via **reference**, no pointers.
- Output matches exactly.

## Expected output
````C++
Email: Welcome aboard!
SMS: Code 1234 confirmed.
````
