# Abstract Classes

## Interfaces

An interface is a contract that states that any object of this type can perform the operations. In C++, we express an interface as an abstract class. You can point to or refer to that abstract class, but you can't create an instance of it. Concrete subclasses fulfill the promise by overriding every pure virtual function. Because callers work through the base pointer or reference, we can swap in new implementations without changing the caller's code. 

In C++, an abstract class is a type that declares behavior without providing the concrete code that implements that behavior. You can make a class abstract by marking one or more of its member functions as pure virtual. By doing so you are stating that the class defines an interface contract that must be met by derived classes that implement the abstract class.

A pure virtual function is declared by following the function declaration with `=0`. That syntax tells the compiler that the function does not have a body, and it forces derived classes to supply one if they are to be instantiated. Any class that still contains an unimplemented pure virtual function stays abstract, so the compiler prevents incomplete objects from being created.

Imagine you're writing a UI toolkit in C++. It might contain buttons, sliders, checkboxes , and text fields. All these UI widgets would need to known how to paint themselves on the screen. Each would draw itself differently. One way to manage this would be to start with an abstract class named widget that declares the operation, every concrete UI widget must support. In this case a draw function. This abstract class would be the base classs for every ohter widget in the UI toolkit.

````C++
class Widget{
    public:
        virtual void draw() const = 0; // pure virtual
        virtual ~Widget() = default;    // C++17, trivial but virtual
};
````

Application code would store and manipulate widgets through pointers or references to the widget class, allowing new widget types to plug in without touching the rest of the system. You wouldn't ever create a plain widget. Instead you would create a sublcasses that knows how to draw itself, marking each override with the override keywork. Once this is done, the subclass can be instantiated and calls to the instantiated object for a base class reference will dispatch to the correct implementation at run time. 

````C++
class Button : public Widget { 
    public:
        void draw() const override { std::cout << "[Button]\n"; }
};
````


An **interface** represents a **contract** -it defines what operations a type must support without specifying how they're implemented. In C++, interfaces are expressed as **abstract classes**. You can hold a **pointer or reference** to an abstract class, but you can't crreate an instance of it directly. 

A class becomes abstract when it declares one or more **pure virtual functions**, marked wiht `=0`. This tells the compiler that the function has **no implementation** in the base class and must be **overriden** by any derived class that wants to be concrete (instantiable).

By programming through **base class references or pointers**, you can **swap implementations** freely - adding new widget types without changing existing code. Once all pure virtual functions are overriden (using the `override keyword`), the derived class becomes concrete and can be instantiated. 

This approach makes your code **extensible, modular, and easy to maintain**, allowing new functionallity to be added without breaking existing systems.
