# Initialization 

In C++ there are two stages of setting an object value, **initialization** and **assingment**.

**Initialization:** is placing a value into an object when the object is first created in the program. 

**Assignment:** is placing a value into an object after it has already been created.

````C++
1| using namespace std; 
2| 
3| string a = "This is initialization";
4| string b; // Initialized to a default value
5| 
6| int main(){
7|    b = "This is assingmentt";
8| }
````

In this case in line 3 is the initialization, the variable `a` is created and in that time it has been assigned the value `"This is initialization"`. In line 4 and 7 is shown the assignment, in line 4 the `b` variable is created, and in line 7 the string `"This is assignment"` is assigned to `b` variable.

In C++, **overloading** is the ability to define multiple functions with the same name, but different parameter lists. This feature also applies tto constructors. The compiler will decide which constructor  to call based on the number and types of parameters provided to the constructor. 

Constructor chaining delgates a portion of the initialization in one constructor to another constructor defined in the class.



In C++, there are several ways to initialize class members: 
-

- **Inline initialization:** Setting default values directly inside the class definition.
- **Default constructors:** Automatically givve member variables initial values when object are created.
- **Initializer lists:** A more efficient way to initialize memebers directly during object creation.
- **Parameter constructor:** Allow you to pass values when different parameter lists let you creatte objects in flexible ways.
- **Constructor chaining:** Lets one constructor call another to reuse initialization logic.

Together, these techniques ensure that your objects always start in a valid state, making your code **cleaner, safer, and easier to maintain**
