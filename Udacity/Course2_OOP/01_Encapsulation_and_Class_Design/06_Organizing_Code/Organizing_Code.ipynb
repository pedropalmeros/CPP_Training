# Organizing Code
## Namespaces

<img src="./IMG/Namespaces_01.png" width="400">

Namespaces are a tool for organizing code into logical groups. As software projects grow, it's common to encounter name conflicts, where two classes or functions have the same name, but are used in different parts of the system. Namespaces help to prevent these crashes by creating named scopes. The C++ standard library itself uses the `STD` namespace, which is why you write `std::cout` and `std::cin`.

````c++
namespace hrlib{
    class Division{
        // Division class definition
    };

    class Employee{
        // Employee class definition
    };

    namespace impl{
        void logEmployeeCreated(const Employee& emp);
    }
}
````
When developing your own libraries or working on large systems, you can define your own namespace to separate unrelated components and clarify where different types belong. For example, an HR library might have a top level hrlib namespace with nested namespace like hrlib implmentation to contain internal details. 

Libraries combine reusable components into binary packages that can be included in a program's library. These can be static library files, which may end in `.lib` or dynamic libraries, which may end in `.dll` or shared libraries, which my end in `.so`. By using libraries, programs do not need to have access to the implementation files for the code compiled into the library. They only need the headers that define the data class, and functions. The code in the libraries is added to the program through the use of a special program called a *linker*. 

<img src="./IMG/Libraries_01.png" width="400">



Starting in C++ 20, the `import` keyword introduces a modern alternative to include for bringing in external code. Traditional include directives, copy the contents of a head of file into your translation unit, which can lead to long compile times, repeated passing, and tricky bugs due to multiple inclusion and hidden dependencies. **Modules**, by contrast, are compiles once and imported wherever they are needed. This improves bill times and enforces a cleaner separation between a module's interface and its implementation. This subsequently leads to more maintainable code, fewer dependencies, and better encapsulation of internal details. 


<img src="./IMG/Modules_01.png" width="400">

- **Libraries** (static `.lib` or `.a`, dynamic `.dll`, or shared `.so`) bundle reusable code that programs can link to
- **Modules** (introduced in C++20) offer a modern, faster alternative to `#include`, reducing compile times and improving encapsulation.
