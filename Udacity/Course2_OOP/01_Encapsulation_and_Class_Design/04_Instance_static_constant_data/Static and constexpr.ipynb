# Static data and member functions

Most of times, in the context of OPP, the data belongs to a specific instance of a class. That means when we create mulitple instances of a type, each instance has its own set of member variables associated with it. Modifying the instace data in one object will not affect the data inother objects.

**Static** data, however, belongs to the class itself and not to a specific instance. It is usually used to track data thtat is related to a class, but which applies to all instances of the class, not just one. ***Static data members  must be defined outside the class definition, unless they are declared inline***. this external definition ensures that there is exactly one definition of the variable in the entire progra. Whithout this, the linker will not be able to resolve the reference to the static member. 

In C++ 17, inline static member variables can also be initialized within the class itself. This simplifies code oganization and avoids the need for out-of-line definitions. 
````c++
class Employee{
    private:
        static inline int instanceCount = 0; 
        string company; 
        string surname; 
        string givenName; 
        };
````


**Static member functions are not tied to any instance of the class, and therefore cannot access non-static members**, thtey are ideal for utility functions that belong to a class conceptually, but do not need to access or modify class state, but do not need to access or modify class state. They are also useful for manipulating private static data.



````C++
class Employee{
    private: 
        static inline int instaceCount = 0; 
        static void outputInstanceCount(); 
}; 

void Employee::outputInstanceCount(){
    cout << "Instance count: " << instanceCount << endl; 
}
````



# constexpr

Constant expressions are expressions whose values are known at compile time. In c++, the `constexpr` keyword enables this behavior by indicating variables or functions whose evaluation must occur during compilation.

In c++, `constexpr` means: "This value or function can be evaluated at compile time". With `constexpr`, the compiler can calculate the result before the program even starts. 

This is helpful for improving performance and ensuring correctness, especially when a value must be used in context where run-time evaluation is not allowed, shuch as array sizes, template parameters, or case labels in switch statements.

Using constant expressions also allows the compiler to perform more aggressive optimizations, as it can substitute fixed values directly into generated code. This is particularly important in systems programming and embedded development, where performance and resourse constraints ar critical. 

````C++
class FileWriter{
    private:
        static constexpr size_t MAX_FILENAME_LENGTH = 255; 
        };
````

The `constexpr` keyword can also be applied to functions to indicate that they can and should be evaluated at compile-time whenever possible. As of C++17, `constexpr` functions are allowed to include control flow like if, for, and switch, greatly increasing their utility. This allows you to define utility functions and even constructors that compute values at compile-time. This leads to a safer and more efficient code base, which is useful in embedded systems, configuration, constants or Meta programming. 

````C++
constexpr int square(int x){
    return x*x; 
}

constexpr int result = square(5);
````
When `constexpr` function is used in the compile-time context, such as initializing another `constexpr` variable, the compiler evaluates the function during compilation. If it's used in the run-time context, then it behaves like a regular inline function.
