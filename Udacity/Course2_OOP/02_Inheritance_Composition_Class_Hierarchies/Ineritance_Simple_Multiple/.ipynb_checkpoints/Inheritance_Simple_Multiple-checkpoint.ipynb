# Inheritance

Inheritance is the pillar of object oriented programming 

<img src="./IMG/Inheritance_01.png" width="400">


Inheritance is the pillar of object oriented programming that allows one class to extend another. It enables the creation of a new class known as *derived* class. That class incorporates the data and behaviors of an existing class known as the *base* class. This promotes code reuse and stablishes **and is a** relation ships. For example, a `car` class might inherit from a more general `vehicle` class, since car is a vehicle.

By using inheritance, programmers can avoid duplicating code in related classes. Common functionality is implemented in the base class, and specialized behavior is added in the derived classes.

When a class inhertis from another, it gains access to the base classes public and protected members, but not its private ones. This access is goverened by the inheritance specifier. 

Public inheritance means the public interface of the base class becomes part of the dervied classes public interface. 

The protected acess specifier allows a class to expose members to its derived classes without making them public. This is useful when basse classes want to share functionality internally with their subclasses, but keep it hidden from external code. 


````c++
class Base{
    protected:
        int id;
};

class Derived : public Base{
public:
    void assingId(int value){
        id = value; 
    }
};
````


When a derived object is created, its base class must be constructed first. If the base class has a default constructor, it will be called automatically. But if the base class constructor requires arguments, then the derived class must explictly call the correct base constructor using an initializer list. 

By centralizing common functionality in the base clas, inheritance helps reduce code duplication and promotes cleaner, more maintainable desing. In C++, **public inheritance** is typically used to make the base class's public interface part of the dervied class's interface - meaning derived objects can be used anywhere a base object is expected. 

**Access specifiers**
- **Public inheritance** exposes the base class's public members.
- **Protected members** are accesible to subclasses but hidden from external code.
- **Private members** remain inaccessible to dervied classes.


In the **constructors the base classs must always be constructed first**. If it has parameters, the derived class must explicitly call the appropriate base constructor using an **initializer list**. 

Together, these concepts from the foundation for **polymorphism** and designing clear, extensible class hierarchies in C++.
