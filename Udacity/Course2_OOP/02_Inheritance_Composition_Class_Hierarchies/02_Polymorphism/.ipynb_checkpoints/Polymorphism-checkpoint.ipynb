# Polymorphism

The word polymorphism meas many forms. In programming, polymorphism refers to the ability of a single interface to represent different underlying forms or data types. In the context of C++ and Object-Oriented programming, polymorphism usually refers to run-time polymorphism. 

**Run-time Polymorphism** is the ability to call the appropriate derived class function through a base class pointer or reference. This is achieved with virtual functions and is often called **dynamic dispatch**.

Polymorphism build on inheritance. With inheritance, we can treat a derived object as if it were a base object. Polymorphism takes this further. It allows a base class to define a function interface, i.e, a virtual function, and derived classes to provide their own implementation of that function. when you call that function on a base pointer or reference, then the program will choose the derived class's override if one exists, making the behavior dependent on the actual object type at run-time. For example, consider a c++ library that implements a graphical user interface. Every visual control might derive from a base class named control. This control class might define apaint function, and each derived class would implement its own version of that method that draws the control onto the screen. There's an endless variety of ways that controls can be put together in an applcation. The GUI libraty will let users create buttons, text boxes, check boxses, and so on. But these would all be drawn through the base control class. The program will decide at runtime how to call the method that implements the right versio of the paint function to draw the right control. the ability to decide which virtual method to call at runtime based on the type of the object creates is called **dynamic dispatch**.

**The dynamic dispatch** is enabled by calling functions for a **v-table**, which is a table of function pointers that maps each virtual function the correct implementation for a specific type. This typically incurs some runtime costs, but the cost is usually minimal and the benefits are significant. For example, imagine we have a pointer to an animal object. At runtime, pointer might actually refer to a dog or a cat or some other subclass. When we call a virtual function like make sound on the animal pointer, the copiler generates code that looks up the function pointer in the object's v-table. This ensures that the correct version of the function gets called, evven though we're using a base class pointer.

<img src="./IMG/Polymorphism_01.png" width="400">

````C++
class Control{
public:
    virtual void draw() const = 0; 
    virtual ~Control() = default;
};

class Button : public Control{
public:
    void draw() const override { std::cout << "[Button]\n"; }
    };

class Slider : public Control{
public:
    void draw() const override { std::cout << "<Slider======>\n"; }
};

void paint(const Control& c) { c.draw(); } // uses dynamic dispatch

int main(){
    std::vector<std::unique_ptr<Control>> ui;
    ui.emplace_back(std::make_unique<Button>());
    ui.emplace_back(std::make_unique<Slider());

    for(const auto& ctrl : ui)paint(*ctrl); // one call site, many forms
}
````



To enable polymorphism in C++, you must make base class functions with the `virtual` keyword. This tells the compiler to use dynamic dispatch when the function is called for a pointer or reference. 

<img src="./IMG/Polymorphism_02.png" width="400">


Derived classes can then override the function to provide their own behavior. Not all functions need to be virtual, only those functions where polymorphic behavior is desired should be marked as virtual. non-virtual functions in a base class cannot be overwritten. 

If you define a non-virtual function in a derived class, it simply hides the base version when accessed through the derived type. **When a derived class supplies its own implementation of a virtual function inherited from a base class, this is known as overriding the base class function**. 

<img src="./IMG/Polymorphism_03.png" width="400">

````C++
class Animal{
public: 
    virtual void makeSound() const{}
};

class Dog : public Animal{
public:
    void makeSound() const override{
        cout << "Wofff!" << endl; 
};
````

Modern C++ provides additional keywords to make polymorphism  safer and intentions clearer. The `override`keyword is used in a derived class's virtual function declaration to indicate that it's intended to override a base class virtual function. If, for some reason, it doesn't match any base class function. For example, you mistyped the function name or it had the wrong signature, then the compiler will produce an error. Using override helps catch errors when you think you're overriding a base function, but you actually aren't. when dealing with polymorphism, one very important aspect is proper cleanup of objects. If you plan to delete objects of derived classes for a pointer to the base class, then the base class should have a virtual destructor. This ensures that the destructor of the derived class is called correctly, preventing resource leaks or undefined behavior. 

**Abstract Class** is a class that htas one or more pure virtual functions.

````C++
class Shape{
public:
    virtual void draw() const =0 ; // presence of a pure virtual makes Shape an abstract interface
    virtual ~Shape() = default;    // virtual destructor for safe polymorphic deletetion
````

**Pure virtual function** ends in equals to zero in its declaration. For example, the draw function in the upper imagein this class makes the class abstract, meaning it cannot be instantiated. Any derived class must provide an implementation for draw if it wants to be a concrete or instantiable class. Using abstract base classes allows you to define interfaces that multiple classes can implement, enabling polymorphism without providing a base implementation.

The `final` keyword can be used in two contexts. one is to indicate that a virtual function cannot be overridden by any further dervied classes. This is useful when you want to close off further customization of a particular function ia subclass, perhaps for design reasons or to allow certain optimizations. The oter use of the `final` keyword is to mark an entire class as final, meaning no other class can inherit from it. If you have a class and you are sure no further duration should occur, for example, it wouldn't make sense to subclass it or maybe if you just want to prevent reuse then you can declare the class as final. 

````C++
class Bird : public Animal{
public:
    void makeSound() const override final{
        cout << "Chirp" << endl; 
    }
};





## APENDIX

### `= default`
In C++, when you see a function declared as `= default`, it means

*Ask the compiler to generate the standard implementation of this function*

This is most commonly used with special member functions such as: 

- Default constructor
- Destructor
- Copy constructor
- Copy assignment operator
- Move constructor
- Move assignment operator

This helps to make the polymorphism explicit.
