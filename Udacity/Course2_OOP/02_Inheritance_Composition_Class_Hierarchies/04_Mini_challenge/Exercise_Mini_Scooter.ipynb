# Exercise

## Task
Model a scooter that **has a** battery. Show to order of construction and destruction via simple prints (no input, no pointers). 

## Requirements
- class Battery
- - Battery (int p) sets percent and prints: Battery constructed (%).
  - int level() const returns percent
  - ~ Battery() prints: Battery destroyed at %
- class Scooter (has-a Battery)
- - Scooter (std::string m, int start) initializes its Battery member and prints: Scooter() constructed
  - void status() const: prints battery %
  - ~Scooter prints: Scooter() destroyed
- In main() (no input): print Start, create Scooter s("Mini",80);, call s.status();, end scope, print Done. 

## Expected Output

Start

Battery constructed (80%)

Scooter(Mini) constructed

MiniBattery: 80%

Scooter(Mini) destroyed

Battery destroyed at 80%

Done
