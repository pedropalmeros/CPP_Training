Inheritance is the pillar of object oriented programming that allows one class to extend another. It enables the creation of a new class known as *derived* class. That class incorporates the data and behaviors of an existing class known as the *base* class. This promotes code reuse and stablishes **and is a** relation ships. For example, a `car` class might inherit from a more general `vehicle` class, since car is a vehicle.

By using inheritance, programmers can avoid duplicating code in related classes. Common functionality is implemented in the base class, and specialized behavior is added in the derived classes.

# Multiple Inheritance

C++ allows a class to inherit from more than one base class. This is known as mulitple inhertiance. It enables us to combine functionality from different sources into a single class, but it must be used with care. 


Inheritance from mulitple sources can introduce ambiguity, specially when the same member appears in morte tha one path. A common issue with mulitple inhertiance is the diampond problem. This happens when tow base clasess both inherit from the same ancestor, and a thrid class inhertis from both of them. 
````c++
class Printer{
    public:
        void print() const;
};

class Scanner{
    public:
        void scan() const;
}; 

class Copier{
    public:
        void scan() const;
};

class AllInOne : public Printer, public Scanner, public Copier {};
````

For example, if you have a base class that defines a vehicle, then a car class and a boat class that derive from vehicle, it's tempting to derive a class for amphibious vehicles from both car and boat. This causes ambiguity, however, should the shared base peer once or twice.
