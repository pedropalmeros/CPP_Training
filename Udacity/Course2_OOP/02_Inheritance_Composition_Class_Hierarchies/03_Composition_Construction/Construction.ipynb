# Construction

During construction, the run-time builds the object from the top down. Virtual bases, if any, are built fist, the non-virtual bases are built in their order of declaration. Finally, the mos derived class is built. A derived constructor chooses which base constructor to invoke via its initializer list. Emitting that choice requress the base constructor to have a default constructor. Object lifetime is a mirror image.

<img src="./IMG/Constructors_01.png" width="400">

If we return to the idea of a hierarchy as a tree, then an object created from such a hierarchy is built from the root upward and destroyed from the leaves downward. Thus, base parts are fully intact while derived constructors run andderived resources are still live when their destructors need them. When using base pointers polymorphically, declare the base class destructors virtual to enshure the chain reverses correctly
