Inheritance is not the only way to reuse code and model relationships between concepts. 

**Composition** is another fundamental design technique, where a class contains instances of other classes rather tan inherit from them. For example, a car class might contain an engine object rather than inherit from engine because a car has an engine, but it's not a type of engine. With composition, the car class can use the public interface of engine by calling public methods on its engine member instead of inheriting all engine's interface. 

Choosing between inheritance and composition depends on the relationship you're modeling. Use inheritance, which models **is a** relationship when the derived class truly is a specialized form off the base clas. For example, a square is a shape. Use composition, which models a **has a** relationship, when a class should be built from one or more other classes, but is not a subtype of those classes. For example, a car has an engine, but a car is not an engine. 

Using composition can make designs mor flexible. For instance, if car has an engine and later, we want to change how engine works, perhaps by replacing it with a different engine type, we can ofter do so without affectir the external interface of car as long as engines interface refamins consistent. Inheritance, changing the base class can ripple into all dervied classes, ptentially requiring changes or recompilation. Inheritance gives a class every public and protected member of its base and signals and `is a` relationship. Composition, by contrast, embeds one object inside another to expres a `has a` relationship. Because the containing class forwards only the operations it wants to expose, composition keeps interfaces tidy and lets parts evolve independently. For example, swap an engine for a new kind without touching the res of the car.

During construction, the run-time build the object from the top down. Virtual bases, if any, are built fist, then non virtual bases are buit in their order of declaration and finally, the most derived class is built. A derived constructor chooses which base constructor to invoke via its initializer list. Omitting that choice requires the base constructor to have a default constructor.  Object lifetime is a mirror image. if we return to the ide of a class hierarchy as a tree, then an object created from such a hierarchy is built from such hierarchy is built from the root upward and destroyed from the leaves downward. Thus, base parts are fully intact while derived constructors run and derived resources are still live when their destructors need them. When using basse pointers polymophically, declare the base class destructors virtual to enshure the chain reverses correctly. 


Derived classes are tightly coupled to their base classes - changing a base (like adding a virtual function or reordering members) can break compatibility or require full recompilation. To avoid such fragility, it's often better to design **base classes for extension** or choose **composition** instead. 

**Composition** means building a class using instances of other clasess rather than inheriting from them. For example, a Car **has an** Engine, but it **is not** a type of engine. The Car class can simply call  the public methods of its Engine member without exposign all of Engine's interface. 

Here's how to decide: 
- Use **inheritance** when there's a true **is-a** relationship
- Use **composition** when modeling a **has-a** relationship

Composition offers greater **flexibility and isolation** - you can swap or modify parts (like replacing an engine type) without affectign the outer class's interface. Inheritance, while useful, can create unwanted coupling between base and derived classes. 

By combining these approaches thoughtfully, you can design systems that are **modular, maintainable, and easy to evolve** over time
