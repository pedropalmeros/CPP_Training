# Runtime polymorphism in C++

In C++ when a function is declared as **virtual** in a base class decides at **runtime** which version of that function to call - depending  on the **actual object type**, not just the pointer type. 

In the example a base class Animal defines a virtal function sound(). The derived classes Dog and Cat **override** this function with their own implementations. When pointer of type Animal point to Dog and Cat objects, calling sound() results in **woof** or **meow**, base on the object they actually refer to. 

This behavior (deciding the correct function at runtime) is what makes polymorphism powerful. It enables **one inteface** (Animal::sound) to support **many implementation**, making programs more **flexible, extensible, and realistic** in modeling real-world behavior. 

````c++
class Animal{
public:
    virtual void sound() { cout << "Some animal sound\n"; }
};
````

On the code, the keyword `virtual` before `void sound`. `virtual` says to c++, when someone calls this function through a **pointer or reference** don't decide now which version to use, wait until runtime and then pick the correct version depending on the actual output.

Without virtual, every call would just use sound from animal.


````c++
class Dog : public Animal{
public: 
    void sound() override {cout << "Woof\n";}
};
````
In the dog class, `public Animal`means dog inherits from animal, It gets all of animals behavior, and it can override functions. the `override` keyword makes it clear we are replacing the base version.
