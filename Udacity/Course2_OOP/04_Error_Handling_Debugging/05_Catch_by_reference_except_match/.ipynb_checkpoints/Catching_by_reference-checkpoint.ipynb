# Catching by Reference and Exception Matching

So far, we've learned how to throw exceptions and then how to catch the. But there are two subtle details that make a big difference in profesional code, how you catch, either by value or by reference, and how the run type matches the throw type with a catch block. 

````C++
catch (const SomeException& e){ .... } 
````

The best practice can be seen in the above snippet code. But you might wonder why that `const`  exception does two different things. 1) It avoids making an extra copy of the exception object and 2) it preserves the dynamic type, if you throw a dervied exception, you don´t accidentally slice it down to its base. 

Think of catching exceptions like handling a fragile package. By value is like asking the delivery person to pen the box, dump the contents into a new box, and hand you that instead, some fragile pieces may get left behind.  

In the same context of fragile packages if the arguments are passed by `const&`, is like just taking the original sealed box directly, no copyingg and no missing pieces.

````C++
try{
    throw std::runtime_error("Runtime problem"); 
}catch(const std::runtime_error& e){
    std::cout << "Good catch: " << e.what() << std::endl; 
}catch(std::runtime_error){
    std::cout << "Bad catch: " << e.what() << std::endl;
}
````

Considering the previous code snippet, in the first block, we bind a reference. No copy is made and we keep the exact type. In the second block, we catch by value. That makes a copy. Fi this were a custom exception with extra fields, like an error code or timestamp, that information might get sliced off. Cathcing by value isn't just slower, it's also lossy. 


How does C++ decide which catch block to use? The runtime looks through the order and the first one that matches wins. 

````C++
try{
    throw std::out_of_range("Index too high"); 
}catch(const st::out_of_range& e){
    std::cout << "Caught out_of_range: " << e.what() << std::endl; 
}catch (const std::exception& e){
    std:;cout << "Caught generic exception: " << e.what() << std::endl;
}
````

Here we throw out-of-range. The runitme starts at the top. The first block matches exactly, so that block runs. If that block weren't there, the second block would still catch it because out-of-range is a sublcass of standard exception. 

````C++
try{
    throw std::invalid_argument("Bad input"); 
}catch( const std::exception& e){
    std::cout << "Generic catch " << e.what() << std::endl; 
}catch( const std::invalid_argument& e){
    std::cout << "Specific catch: " << e.what() << std::endl;
}
````
Here, we throw an invalid argument. But look at the order. The generic standard exception handler comes first. That block matches everything derived from standard exception, so it runs immediately. The more specific invalid argument handler is never reached. This is why order matters.

Always list specific catches first and the general safety net last. Here are the ruls of thumb.
- Catch by `const &` to avoid copies and preserve type.
- Exception matching is type-based.
- The first match wins.
