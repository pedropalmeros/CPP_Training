# The `Throw` keyword

`Throw` keyword is how hwe signal that an error has occurred. Thik of it like pulling the emergeny brake on a train. Everything stops and control is passed to whoever is responsible for handling the problem.

Technically, C++ will let you throw almost anything. Integers, strings, even custom objects, but in professional code, we always throw exceptions derived from `std::exception`. That way, all errors share a common interface, i.e, the what method and can be caught consistently.

````C++
void bad_function(){
    throw 42; // Legal, but meaningless
}
````
The code in the previous snippet works, compiles, but it's useless. What does 42 mean? Without documentation, the caller has no idea whether it's a file rror, a math error, or something else.


````C++
void also_bad(){
    throw "File not found"; // Slightly better, but not standard
}
````
The code in the previous snippet at least gives a human readable message, but it's still not part of the standard exception hierarchy. The caller would need a separate catch just for strings. It won't be caught by a normal catch.




````C++
#include <fstream>
#include <stdexcept> 
#include <string> 

std::string read_file(const std::string& path){
    std::ifstream file(path);
    if(!file){
        throw std::runtime_error("Cannot open file: " + path); 
    }
    return std::string((std::istreambuf_iterator<char>(file)), std::istreambuf_iterator<char>());
}
````
In the previous snippet, if the file is not found the code throws a `std::runtime_error` with a clear message. This integrates perfectly with the standard hierarchy, so the caller can simply use a catch to handle it. **The rule of thumb is throw early, throw clearly, and throw something derived from `std::exception`.**


    

Sometimes you want to promise that a function will never trow an exception, That's where no except comes in.  Marking a function, no except tells both the compiler and other developers, this function will not throw exceptions. 
````C++
void safe_function() noexcept{
    // Guarandeed not to throw
}
````
Why does `noexept` is important? Firstly, for performance. the compiler can generate faster code if it knows, exceptions won't escape. next, for clarity. it documents you intent. Other programmers immediately know this function is guranteed not to throw exceptions, so they don't waste time wrappping it in unnecessary try catch blocks. Finally for safet in destructors, by default, destructors are implicitly no except true. This is important because if a destructor throw while another exception is already being handled, the program calls standard terminate. But there's a catch. If a no accept function does throw, the program immediately calls `std::terminate`. No stack unwinding, no chance to recover, it's fatal. You should only use no accept when you're absolutely certain the function shouldn't throw exceptions. The `throw` keyword is your emergency brake. It signals that an error has occured. No except is your contract. It guarantees a function won't throw an exception at all ever. Together, they give you both sides of the ver handling story, when to signal problemas and when to guarantee safety. 


Here we learn how **C++ signals and controls exceptions** using `throw` and `noexcept`.

The **`throw` keyword** is like an emergency brake - it stops normal execution and signals that an error has occurred. Technically, you can throw anything (like integers or strings), but best practice is to **trow objects dervied from `std::exception`, ensuring a consistent interface through the what() method and simpler, unified error handling. 

---
**The golden rule**
- **Throw early**
- **Throw clearly**
- **Throw something dervied from `std::exception`**
---

On the other hand, the **noexcept keyword** promises that a function will **never throw exceptions**. This improves **performance**, provides **clarity** for developers, and is essential in **destructors**, since throwing during stack unwinding would trigger `std::terminate`. 

Together, `throw` and `noexcept` from the two sides of exeption handling - **one for signaling problems**, and **one for guaranteeing saffety and dreictability**.
