We have three different tools in our toolikt: **exceptions**, **return codes** and **optionals**. Each one has strengths, and weaknesses. 

**Exceptions** are perfect for rare, serious problems that break normal flow. If a configuration file is missing, that's not a normal state. You expect it to be there. If a network suddenly goes down, that's unexpected. If memory allocation fails, that's a critical error. These are the fire alarm situations. You don't expect them to happen often, but hwne the do, they interrupteverything, and must be handled. 

**Return codes** are simple and efficient. They're best when the function is called millions of times per secdon and perfromance matters. The outcome is just success or failure with maybe a small detail about what went wrong. The absence of a result is not an error, it's a valid state. This is like the waiter saying sorry, no ketchup. It's part of the normal flow. No alarms needed, just a quick signal and move on. 

**`std::option<T>`** is perfect when a value may or may not exist. A search in a list, looking up a key in a config map, passing user input hat might be missing. These are gift box cases. Sometimes you open the box and there's a present inside, sometimes it's empty. Both outcomes are fine. The optional forces you to check before using the value, making it safer than sentinels like minus one or empty strings.
