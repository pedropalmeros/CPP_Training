# Try and Catch

Exceptions are only useful when actually handle them. Otherwise, they're just flares with nobody watching the sky. That's where the `try`and `catch` keywords come in. A `try`block wraps code that might throw an exception. Then, one or more `catch` blocks follow, each describig how to handle a specific type of exception. ````C++
int main(){
    try{
        std::string config = read_file("config.txt"); 
        int port = parse_port("8080"); 
        std::cout << "Port: " << port << std::endl; 
    }catch(const std::invalid_argument& e){
        std::cerr << "Invalid input: " << e.what() << std::endl; 
    }catch(const std::out_of_range& e){
        std::cerr << "Out of range: " << e.what() << std::endl; 
    }catch(const std::runtime_error& e){
        std::cerr << "Runtime error: "<< e.what() << std::endl; 
    }catch(const std:;exception& e){
        std::cerr << "Unexpected error: " << e.what() << std::endl; 
    }
}

Inside the `try` block , we frist attempt to read a convict file. If that fails, refiled, throws a runtime error. Next, we pass a port number. If the string is invalid parse_port throws invalid argument. If the number is too big, it throws out of range. Matching with cat blocks. If the error is sepecifically invalid argument, the frist catch block runs, printing invalid input. If it's runtime error, the third block handdles it. Finally, the catch block is a safety net. It can catch anything else in the standard exception hierarchy. Orders matters. The program checks the catch blocks from top to bottom. If we put the generic catch first, it would just grab everything and the more specific handlers would never run. That's why we list the specific catches first and the most general last. Think of it like a hospital triage. The first doctor specializes in heart problems. The second doctor specializes in broken bones. The third doctor specializes in infections. Finally there's a general doctor who sees any patient that doesn't fit the others. If you put the general doctor first, the specialists woud never get a chance. Catch ordering works in the same way. Here's the pattern to remember. Wrap risky code in try block, handle known problems with specific catch blocks, and always include a final catch as a safety net. Remember, order matters, specific first, general last. That way, your program doesn't just crash when something goes wrong. It responds intelligently, like a professional system should. 

Now we explore how to **handle exceptions properly** using try and catch. 

A **try block** wraps code that might fail, while one or more **catch blocks** define how to respond to specific types. Each catch acts like a **specialized safety net** -for example, one might handle `std::invalid_argument`, another `std::out_of_range`, and a final one catches general `std::exception` errors. 

**Order matters**. The program checks catch blocks **from top to bottom**, running the first one that matches. Always list **specific handlers first** and **general ones last**, so the right error gets handled properly. 

The pattern to remember:
- Wrap risky code in try.
- Handle expected errors with targeted catch blocks.
- End with a generic catcha s a final fallback.

This ensures your program doesn't just fail - it **recovers gracefully** and responds intelligently to whatever goes wrong.
