The moment you throw an exception, execution of the current function stops immediately. No more lines in that function are run. The program starts searching up the call stack, that's the trail of functions that led here for a catch block that matches the type of exception. As it searches the stack unwinds, local objects go out of scope, their destructors are called automatically, and resources like memory, files, or sockets are cleaned up along the way. If it finds a matching cat, that block runs. If no one catches it, the program calls std::terminate and ends.

Now we explore how **exceptions** bring safety and structure to error handling in C++. 

You learned that an **exception** is an **object representing an error or unusual condition**, allowing programs to respond consistently without crashing unpredictably. 

When an exception is **thrown**, exceution of the current stops, and C++ begins **stack undinding**-automatically calling destructors and releasing resources as it searches for a matching **catch block**. If no handler is found, the program terminates cleanly instead of failing silently. 

This mechanism separates **error detection** from **error handling**-the function that detects a problem doesn't need to fix it. Instead, the caller decides how to respond. 

By using **throw, try,** and **catch**, we can signal problems clearly, clean up safely, and build programs that fail predictably instead of dangerously.
