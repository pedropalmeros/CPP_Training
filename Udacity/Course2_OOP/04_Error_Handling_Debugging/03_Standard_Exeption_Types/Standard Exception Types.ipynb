# Standard Exception Types

In C++, not all errors are equal. Some are programmer mistakes, some are bad input. Some are system failures. That's why the standard library givs us a family of exception types under standard exception. Choosing the right one makes your code self documenting and predictable. 


---
## `std::runtime_error`

````C++
#include <fstream>
#include <stdexcept>

void read_file(const std::string& path){
    std::ifsetream file(path);
    if(!file){
        throw std::runtime_error("File not found: " + path);
    }
}
````
The previous snippet code is used when there are unexpected failures at runtime, like missing files. Here we try to open a file. If it doesn't exist, we throw a standard runtime error. This type is the workhorse of runtime problems. Thinks like network failures, missing resources or failed IO, it says, this isn't the caller's fault, but something went wrong at runtime. 

---


## `std::invalid_argument`

````C++
#include <stexcept>

int square_root(int n){
    if(n < 0){
        throw std::invalid_argument("Square root of negative not allowed"); 
    }
    // fake sqrt for demo

    return n/2; 
}
````

Here, the caller can pass a negative number to a function that only makes sense for positives. That's not a runtime failure. It's invalid input. Throwing invalued arguments makes it clear the function was misused. If you're writing library code, this exception type is gold. It tells the user when they've called something the wrong way.

---

## `std::out_of_range`

````C++
#include <vector> 
#include <stdexcept>

int get_element(const std::vector<int>& v, int index){
    if(index < 0 || index >= static_cast<int>(v.size())){
        throw std::out_of_range("index is outside the vector bounds"); 
    }
    return v[index]; 
}
````

Here, if the caller asks for element then in a vector of size three, we throw out of range. This prevents undefined behavior. Out of range is great for bounds checking things like strings, arrays, or containers. Basically, anywhere you want to say that index doesn't exist.

---

## `std::logic_error`

````C++
#include <stdexcept> 
int divide(int a, int b){
    if(b == 0){
        throw std::logic_error("Division by zero"); 
    }
    return a / b; 
}
````
In case of violated assumptions or programming mistakes, we use logic error. Divison by zero isn'treally a runtime failure. It's a logic mistake. The program should never even try this. Here we throw logic error. It communicates the code itself isn't wrong. The function was called in a way that violates its contract. Use this when a condition should never happen if the program is written correctly. 

---

## `std::bad_alloc`

````C++
#include <new> // for std::bad_alloc
#include <iostream>

int main(){
    try{
        // Try to allocate an impossibly large array
        int* huge = new int[100000000000000000000000000000000000000ULL]; 
        delete[] huge;
    }catch (const std::bad_alloc& e){
        std::cout<<"Allocation failed: " << e.what() << std::endl; 
    }
}
````

This exception is used when the moemory allocation fails. This one is special. If the system can't allocate memory, then new froze bad allocation. It's rare in everyday programs, but in memory limited environments or when you request something absurdly large, it can happen. `std::bad_alloc` says the system itself ran out of resources.


# Custom Exceptions

Sometimes the standard exceptions aren't quite enough. Imagine you're building a large application with dozen of different scenarios. Think things like file issues, database issues, configuration issues, or network issues. If you only eer throw a standard runtime error, it's like a doctor telling you you're sick, but without telling you whether it's cold, a broken bone or food poisoning. It's technically true, but it's not really that helpful. This is where custom exceptions come in. You can create your own exception types that make the error crystal clear. The nice thing is, you still build them on top of a standard exception. Tye integrate perfectlly with the C++ error handling system. 

---
Example
````C++
#include <stdexcept>
#include <string> 
class ConfigError : public std::runtime_error{
public:
    explicit ConfigError(const std::string& msg): std::runtime_error("Config error: " + msg){}
};
````
---

Here, we define config error, which inherirts from standard runtime error. Now you might wonder why inherits, because then config error is still part of the exception family, so it can be caught with catch. In the constructor, we add config error before the message. This way , every thrown error makes it clear that this is a configuration problem.

Think of exception types as labels on boxes in a warehouse. If every box is just labeled error, then workers have no clue where to put anything or how to deal with it. But if some boxes are labeled keep refrigerated or fraginle gass, the handlers know exactly what to do. Custom exceptions are just those more specific labels. they make your programs errors easier to sort, handle, and understand. 

Use standard exceptions when they fit, they cover most cases. But when you need more precision, define a custom exception. It's like putting a clear label on the box. Whoever catches it knows exactly what went wrong and how to then respond. 

Here we focus on **being precise with exceptions** - because not all errors are the same .

In C++, different problems call for different exceptions types. 

- **std::runtime_erorr** is used for unexpected runtime failures, like missing files or network issues.
- **std::invalid_argument** signals bad input or incrrect API use.
- **std::out_of_range** prevents undefined behavior from invalid indexes.
- **std::logic_error** flags programming mistakes, such as dividing by zero.
- **std::bad_alloc** indicates that the system ran out of memory.


Choosing the **right exception type** makes your code self-explanatory and helps other respond correctly. 

When built-in types aren't specific enough, you can create **custom exceptions** by inheriting from `std::exception` (often `std::runtime_error). This allows you to define clear, meaningful categories like ConfigError or DatabaseError, while still fitting into C++'s exception-handling system. 

In short,think of exception types as **labels**- the clearer they are, the easier is for you program (and other developers) to know exactly **what went wrong** and **how to handle it**
