# Stack Unwinding to RAII 

One of the most powerful things about exceptions in C++ is how they work hand in hand with RAII (Resource Acquisition Is Initialization). the rule is simple but profound. Whenever an exception is thrown, C++ unwinds the stack, and during that process, it automatically calls destructors for every local object that has gone out of scope. That means, even if your program fails halfway through, resources like files, memory, or sockets all still get cleaned up. You don't need to remember every exit path. The comiler does it for you.

````C++
#include <iostream>
#include <stdexcept> 
class Tracer{
public:
    Tracer(const std::string& name) : name_(name){
        std::cout <<"Constructing" << name_ << std::endl; 
    }
    ~Tracer(){
        std::coutt << "Destroying " << name_ << std::endl; 
    }
private:
    std::string name_;
};

void demo(){
    Tracer a("A"); 
    Tracer b("B"); 
    throw std::runtime_error("Something went wrong"); 
    Tracer c("C"); 
}

int main(){
    try{
        demo();
    }catch (const std::exception& e){
        std::cout << "Caught: " << e.what() << std::endl; 
    }
}
````

In the previous coe, when `demo()` starts, we constuct `A` and `B`, then a runtime error is thrown. At tat point, exceution of the function stops immediately. Observe that object `C` is never constructed because code after the throw is skipped. But before leaving the function, destructors run in reverse order of construction. First, `B` is destroyed, then `A`. Think of a hotel with row of rooms. Every time you enter a room, you leave your back there. Now, if a fire alarm goes off, the exception you don't calmly walk forward to the next rooms, you rush back out, grabbin your bans in reverse order as you leave. That's exactly what C++ does during stack unwinding. Each destructor cleans up its bag before the program escapes the function. This tiny example that we saw just prints messages. But in real life, those destructors migght close a file, release memory, unlock a mutes, or disconnect a socket. 

That's the heart of RAII. You tie a resources lifetime to an object. When the object goes out of scope, whether normally or because of an exception, the resource is automatically cleaned up. Stack unwinding and RAII together, give you bulletproof cleanup. You don't need to write cleanup code in every error path. Just wrap resources in objects and let destructors do their job. Exceptions don't leak resources, they actually make clean up atomatic.

- **Exceptions represent errors**
- **`throw` halts normal flow**
- **Use standard exception types for clarity**
- **`try` and `catch` blocks handle errors gracefully.**
- **Use `const&`to avoid copies and slicing.**
- **RAII ensures automatic cleanup through stack unwinding**
---


Here we see how **exceptions and RAII (Resource Acquisition Is Initialization) work together to make C++ code safe and self-cleaning. 

When exception is thrown, **stack unwinding** begins - C++ automatically calls the destructors of all local objects that were successfully created, in **reverse order of construction**. This guarantees that resources like files, memory, and sockets are properly released even when the program exits early due to an error. 

this principle is the core or **RAII**: tie each resource to an object's lifetime. When the object goes out of scope- whether normally or because of an exception - its destructor performs cleanup automatically. 

In short: 
- **Exceptions signal errors** using throw.
- **`try` and `catch`** handle them safely and predictably.
- **Catching by const** avoids slicing and copies.
- **RAII + exceptions** ensure automatic, reliable cleanup through stack unwinding.

Together, these make C++ error handling **robust, resource-safe, and professional-grade**. No manual cleaunp or fragile error paths required.
