# `std::optional<T>`

Think of `optional` as box. Sometimes the bos has a value inside, ando sometimes it's empty. If it has something, you can open it and use it. But if it's emtpy, that just means *no result*. This is perfect for situations where failure isn't catastrophic. It's not fire alarm territory, it's maybe yes, maybe no. Some examples are 

- Searching for a character in a string,
- Looking up a config key,
- Maybe passing optinal input.

In the past, we used return codes with `-1`as a sentinel or just empty strings. But those were easy to misuse and not very clear. **`optional`** makes the intent explicit. 

````C++
#include <optional>
#include <string> 

std::optional<int> find_index(const std::string& text, char c){
    for(size_t i = 0; i < text.size(); i++){
        if(text[i] == c) return static_cast<int>(i);
    }
    return std::nullopt; // no match
}
````
In the previous snippet code, we search through a string. If we find the character, we return its position wrapped in an optional `int`. If we don't, we return standard null option. It's like saying this box might have an index inside, or it might be empty. No exceptions, no magic numbers. Just a clear modern way to say, maybe there's a value, maybe not. 

At the core site, the code is crystal clear. (main function, code below)
````C++
int main(){
    auto pos = find_index("hello",'e'); 
    if(pos){
        std::cout << "Found at index " << *pos << std::endl; 
    }else{
        std::cout << "Character not found" << std::endl; 
    }
}
````
We check if (pos), if the optional has a value, it behaves like true. To get the value, we use `*pos` to unwrap it. If it doesn't exist, we handle the not found case gracefully. Here's the key. The compiler forces you to check. You can't accidentally treat pos like a plin `int` without unwrapping it

## `[[nodiscard]]`
Nodiscrad is a c compilier attribute introduce in C++17. it tells the compiler if someone calls this function and ignores the return value, issue a warning. The goal is to prevent suble bugs where you silently throw away important results. Think of it as a post in note for functions that says, don't ignore me. Without nodicard, it's easy to write code that looks fine, but does nothing useful. 

````C++
[[nodiscard]] std::optional<int> find_index(const std::string& text, char c);
find_index("hello", 'e'); // compiler warns: result is discarded
````
This is useful when 
- Success or failure(bool, error codes, optional).
- Allocate / manage resources
- Ignoring result is a bug.
